```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(context="paper", style="ticks")
```

```python
df = pd.read_csv("data/ewc_hpt_results.csv")
```

```python
df
```

```python
x_col = "WER AMI (%)"
y_col = "WER MLS French (%)"

sns.scatterplot(data=df.astype({"lambda": "category"}), x=x_col, y=y_col, hue="lambda")

plt.tight_layout()
plt.savefig("pairplot_wer_ewc.png")
```

```python
fig, axis = plt.subplots(1, 2, figsize=(8, 3))
df.plot.scatter(x="lambda", y="WER AMI (%)", logx=True, ax=axis[0])
df.plot.scatter(x="lambda", y="WER MLS French (%)", logx=True, ax=axis[1])
fig.tight_layout()
fig.savefig("wer_wrt_lambda_ewc.png")
```
