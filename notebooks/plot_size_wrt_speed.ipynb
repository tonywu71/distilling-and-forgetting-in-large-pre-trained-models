```python
import pandas as pd
from scipy.stats import pearsonr

import seaborn as sns
```

```python
df = pd.read_csv("data.csv")

df
```

```python
sns.lmplot(data=df, x="Size (M parameters)", y="Average latency (ms)", ci=None)
```

```python
# Compute the R2 score using scipy.stats:
r, p = pearsonr(df["Size (M parameters)"], df["Average latency (ms)"])
r
```
